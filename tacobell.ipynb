```python
import pickle
with open('stanford_output.txt', 'rb') as f:
    stanford_output = pickle.load(f)
```

```python
from dataset import get_treebank_3914
indices = get_treebank_3914()['test_ids']
truth = get_treebank_3914()['test_tags']
```

```python
truth_tups = []
stanford_tups = []
for i in range(len(indices)):
    print(truth[i])
    print(stanford_output[indices[i]])
    print()
    truth_tups += truth[i]
    stanford_tups += stanford_output[indices[i]]
```

```python
correct = 0
incorrect = 0
truth_keys = {}
stanford_keys = {}
for i in range(len(truth_tups)):
    truth_tup = truth_tups[i]
    stanford_tup = stanford_tups[i]
    assert(truth_tup[0] == stanford_tup[0])
    truth_keys[truth_tup[1]] = True
    stanford_keys[stanford_tup[1]] = True
    if truth_tup[1] == stanford_tup[1]:
        correct += 1
    else:
        incorrect += 1
assert(correct + incorrect == len(truth_tups))
print(f'correct: {correct}/{len(truth_tups)} ({100*correct/len(truth_tups)}%), incorrect: {incorrect}/{len(truth_tups)} ({100*incorrect/len(truth_tups)}%)')
```

```python
print(set(list(truth_keys.keys())) - set(list(stanford_keys.keys())))
print(set(list(stanford_keys.keys())) - set(list(truth_keys.keys())))
```

```python
correct = 0
incorrect = 0
skipped = 0
for i in range(len(truth_tups)):
    truth_tup = truth_tups[i]
    stanford_tup = stanford_tups[i]
    assert(truth_tup[0] == stanford_tup[0])
    if truth_tup[1] in ['-NONE-', '#', '-LRB-', '-RRB-']:
        skipped += 1
        continue
    if truth_tup[1] == stanford_tup[1]:
        correct += 1
    else:
        incorrect += 1
assert(correct + incorrect + skipped == len(truth_tups))
print(f'skipped: {skipped}/{len(truth_tups)} ({100*skipped/len(truth_tups)}%)')
print(f'correct: {correct}/{(correct + incorrect)} ({100*correct/(correct + incorrect)}%), incorrect: {incorrect}/{(correct + incorrect)} ({100*incorrect/(correct + incorrect)}%)')
```
